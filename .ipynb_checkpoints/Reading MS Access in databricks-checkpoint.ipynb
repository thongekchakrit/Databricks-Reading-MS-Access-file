# Reading accdb on databricks

#credits: https://gist.github.com/guanjieshen/0731947efdca5356a322f5de1e164b46

On top of his code, I've added a further step showing which dependencies are needed for Ucanaccess to work on databricks.

Without the steps errors such as "java.lang.NoClassDefFoundError: com/healthmarketscience/jackcess/util/ErrorHandler" occurred. And without install uncanaccess jar file in cluster libraries manager, databricks was unable to point to Ucanaccess driver.

### Step # 1

Download UcanAccess driver and dependencies from "https://jar-download.com/?search_box=ucanaccess"

Install all driver and dependencies using Cluster Library manager

### Step # 2
Download a copy of accdb file into databricks

```python
%sh
wget https://download.microsoft.com/download/e/9/8/e981203a-d902-4d63-afbf-424027b1e88c/olympicmedals.accdb
mkdir -p /dbfs/FileStore/accessdb/
mv olympicmedals.accdb /dbfs/FileStore/accessdb/
```

### Step # 3
Read accdb file and save as a dataframe

```python
connectionProperties = {
  "driver" : "net.ucanaccess.jdbc.UcanaccessDriver"
}

url = "jdbc:ucanaccess:///dbfs/FileStore/accessdb/olympicmedals.accdb"
df = spark.read.jdbc(url=url, table="MEDALS", properties=connectionProperties)

display(df)
```

# Step #4
Save dataframe as tempt table so it can be used from SQL

```python
%python
df.createOrReplaceTempView("data_metals")
```

```python
%sql 
SELECT 
  Athlete, Medal, COUNT(Medal)
FROM data_metals
GROUP BY Athlete, Medal
ORDER BY Medal
```
